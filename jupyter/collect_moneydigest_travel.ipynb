```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

```python
# Collect travel deals from 1 page
category_link = "http://www.moneydigest.sg/category/deals/travel-deals/"
category_resp = requests.get(category_link)
category_soup = BeautifulSoup(category_resp.text, 'html.parser')
```

```python
articles = category_soup.find_all('article')
for atc in articles:
    h2_node = atc.find('h2')
    h2__a_node = h2_node.find('a')
    print(h2_node.text)
    print(h2__a_node['href'])
    print()
```

```python
# Collect details from article
article_link = articles[0].find('h2').find('a')['href']
article_resp = requests.get(article_link)
article_soup = BeautifulSoup(article_resp.text, 'html.parser')
```

```python
article_node = article_soup.find('article')

article_title = article_node.find('h1').text
print(article_title)

article_text = article_node.find_all('p')
for p in article_text:
    print(p.text)
```

```python
tags_node = article_node.find('div', attrs={'class' : 'entry-tags'})
tags__a_nodes = tags_node.find_all('a')
for t in tags__a_nodes:
    print(t.text)
```
